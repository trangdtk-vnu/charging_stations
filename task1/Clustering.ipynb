```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

top 10 towns

```python
features = [
    'AvailableSurroundingChargers_Count', 
    'LastSignOfLife_Month', 
    'LastSignOfLife_Year', 
    'cableLength', 
    'carChargeParameter.current', 
    'carChargeParameter.maxPower'
]
data_clean = df_final[features].dropna()
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_clean)

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

optimal_k = 3  # You can change this after reviewing the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

data_clean['Cluster'] = clusters

plt.figure(figsize=(8, 5))
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters, cmap='viridis')
plt.title('K-means Clustering')
plt.xlabel('Feature 1: AvailableSurroundingChargers_Count')
plt.ylabel('Feature 2: LastSignOfLife_Month')
plt.grid(True)
plt.show()
```

Brand vs car model vs location


```python
car_column = 'carChargeParameter.car'  
town_column = 'locationProvince'  
charger_brand_column = 'station_brand'

df_encoded = pd.get_dummies(df_final[[car_column, town_column, charger_brand_column]])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

optimal_k = 3 
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

df_final['Cluster'] = clusters

relationship_data = df_final.groupby([car_column, town_column, charger_brand_column, 'Cluster']).size().reset_index(name='count')

pivot_table = relationship_data.pivot_table(values='count', index=[car_column, town_column], columns=charger_brand_column, fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Relationship between Car Types, Towns, and Charger Brands with Clusters')
plt.xlabel('Charger Brand')
plt.ylabel('Car Type and Town')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

return users


```python
evID_column = 'evId'  
session_column = 'session.chargingSessionId'  

user_sessions = df_final.groupby(evID_column)[session_column].nunique().reset_index(name='session_count')

plt.figure(figsize=(10, 6))
plt.hist(user_sessions['session_count'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Number of Sessions per User')
plt.xlabel('Number of Sessions')
plt.ylabel('Number of Users')
plt.grid(True)
plt.show()

repeat_users = user_sessions[user_sessions['session_count'] > 1]
non_repeat_users = user_sessions[user_sessions['session_count'] == 1]

print(f"Number of repeat users: {len(repeat_users)}")
print(f"Number of non-repeat users: {len(non_repeat_users)}")

clustering_data = user_sessions[['session_count']]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

optimal_k = 3  # Adjust this value based on the elbow plot
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
user_sessions['Cluster'] = kmeans.fit_predict(scaled_data)


plt.figure(figsize=(10, 6))
plt.scatter(user_sessions['session_count'], user_sessions['Cluster'], c=user_sessions['Cluster'], cmap='viridis')
plt.title('K-means Clustering of User Return Rate (Number of Sessions)')
plt.xlabel('Number of Sessions')
plt.ylabel('Cluster')
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()

cluster_summary = user_sessions.groupby('Cluster')['session_count'].agg(['count', 'mean']).reset_index()
print(cluster_summary)
```

Polular time


```python

start_time_column = 'session.start'  

df_final[start_time_column] = pd.to_datetime(df_final[start_time_column])

df_final['Month'] = df_final[start_time_column].dt.month
df_final['Hour'] = df_final[start_time_column].dt.hour

# Define seasons based on month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df_final['Season'] = df_final['Month'].apply(get_season)

month_popularity = df_final.groupby('Month').size().reset_index(name='count')
season_popularity = df_final.groupby('Season').size().reset_index(name='count')
hour_popularity = df_final.groupby('Hour').size().reset_index(name='count')

# Most popular month
plt.figure(figsize=(10, 6))
plt.bar(month_popularity['Month'], month_popularity['count'], color='skyblue')
plt.title('Most Popular Month for Charging')
plt.xlabel('Month')
plt.ylabel('Number of Sessions')
plt.grid(True)
plt.show()

#  Most popular season
plt.figure(figsize=(10, 6))
plt.bar(season_popularity['Season'], season_popularity['count'], color='lightgreen')
plt.title('Most Popular Season for Charging')
plt.xlabel('Season')
plt.ylabel('Number of Sessions')
plt.grid(True)
plt.show()

#  Most popular hour
plt.figure(figsize=(10, 6))
plt.bar(hour_popularity['Hour'], hour_popularity['count'], color='salmon')
plt.title('Most Popular Hour of the Day for Charging')
plt.xlabel('Hour of the Day (24-hour format)')
plt.ylabel('Number of Sessions')
plt.grid(True)
plt.show()

df_final['Season_numeric'] = df_final['Season'].map({'Winter': 0, 'Spring': 1, 'Summer': 2, 'Fall': 3})

# Use Month, Hour, and Season (numeric) for clustering
clustering_data = df_final[['Month', 'Hour', 'Season_numeric']].dropna()

scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

```
